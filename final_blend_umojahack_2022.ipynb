```python
import pandas as pd
import numpy as np
```

```python
weights = [0.15405053, 0.11499462, 0.27805919, 0.15806995, 0.29482571]
experiments = ["exp_v4/results/exp_v4.csv",
               "exp_v5/results/exp_v5.csv",
               "exp_v6/results/exp_v6.csv",
               "exp_v7/results/exp_v7.csv",
               "exp_v11/results/exp_v11.csv",               
              ]
```

```python
subs_list=experiments
```

```python
pred_list = []
for sub in subs_list:
  print(sub)
  df = pd.read_csv(sub)
  pred_list.append(np.array(df.Signal))

test_pred=np.average(pred_list , axis=0 , weights=weights)
test=pd.read_csv(subs_list[0])
sub=test[["ID"]].copy()
sub["Signal"]=test_pred
sub.to_csv("final_submission.csv",index=False)
```
